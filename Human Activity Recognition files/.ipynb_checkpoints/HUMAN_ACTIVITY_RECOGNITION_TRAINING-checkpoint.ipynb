# ***mention Training and testing path***

```python
train_path = '/content/Training'
test_path = '/content/val'
```

# ***Apply Data Augmentation***

```python
batch_size=16
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
train_datagen = ImageDataGenerator(rescale=1./255 , shear_range=0.2 , zoom_range=0.2 , horizontal_flip=True,vertical_flip=True)

test_datagen = ImageDataGenerator(rescale=1./255)
```

```python
train_set = train_datagen.flow_from_directory(train_path,target_size=(100,100),class_mode='categorical',batch_size=batch_size)

test_set = test_datagen.flow_from_directory(test_path,target_size=(100,100),class_mode='categorical',batch_size=batch_size)
```

```python
test_set.class_indices
```

```python
train_set.class_indices
```

# ***import library for train the model***

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
from keras.layers import Conv2D, MaxPool2D, Dense, Flatten, Dropout
from tensorflow.keras.applications.vgg16 import preprocess_input,VGG16
import tensorflow as tf
from tensorflow.keras.models import  Model
from tensorflow.keras.preprocessing import image
```

# ***Define Vgg16 model***

```python

vgg = VGG16(include_top=False,weights="imagenet",input_shape=(100,100,3))
```

# ***Don't need to trained again hidden layersa***

```python
for layer in vgg.layers:
  layer.trainable=False
```

# ***Flattening output***

```python
x = Flatten()(vgg.output)
```

# ***Output layer with 45 classes***

```python
x = Dense(45, activation="softmax")(x)
```

```python
model = Model(inputs=vgg.input, outputs=x)
```

# **Compile model**

```python
model.compile(optimizer='adam',loss='categorical_crossentropy', metrics=['accuracy'])
```

# ***Train model for 70 epochs***

```python
epochs = 70
history = model.fit(train_set, batch_size=16, 
                    epochs=epochs, 
                    validation_data=test_set,
                    steps_per_epoch=len(train_set)//3,
                    validation_steps=(len(test_set))//3
                    ) 
```

# ***Evaluation***

```python
score = model.evaluate(test_set)
print('Test accuracy:', score[1])
```

# ***Visualize training and validation accuracy***

```python
# accuracy
plt.figure(0)
plt.plot(history.history['accuracy'], label='training accuracy')
plt.plot(history.history['val_accuracy'], label='val accuracy')
plt.title('Accuracy')
plt.xlabel('epochs')
plt.ylabel('accuracy')
plt.legend()
plt.show()
```

# ***Visualize training and validation loss***

```python
# Loss
plt.plot(history.history['loss'], label='training loss')
plt.plot(history.history['val_loss'], label='val loss')
plt.title('Loss')
plt.xlabel('epochs')
plt.ylabel('loss')
plt.legend()
plt.show()
```
